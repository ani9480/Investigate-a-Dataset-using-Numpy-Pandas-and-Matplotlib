## b. Filtering the dataset 

#### Finding the top 10 Teams of last 2 decades.

#### While exploring each table we can see that there are some fields which are useful for my analysis while some are not. So I will be creating views of tables with fields i need for my analysis. Here first of all i will be deciding the threshold value for year ie. between how many years i will be analyzing my dataset. In this project i am looking at the stats for batters and pitchers only not the fielders.Here are steps for finding out the top 10 teams.

#### 1. So first I filtered the whole data to take entries from the year 1990 till current date. Basically i want to take the data from last 2 decades. 

#### 2. Next I found the teams which had won the world series and leagues in that interval.

#### 3. After that i combined both league data and world series data to find a combined stat of all the teams.

#### 4. Finally i sorted the data according to most world series and league wins and found out the top ten teams in the current time interval.

#### 5. Now for the rest of the exploration purpose i will be looking at the players of these top 10 teams only. By this way i will get good idea about the performance of player , get to know the top players in baseball, their salary stats etc.




```python
##Filtering the data for taking team stats from the last 2 decades
new_team_data = team_data[(team_data['yearID'] >= 1990)]


##Finding the teams which has won the league or world series

leaguewin_teams_data = (new_team_data[(new_team_data['LgWin'] == 'Y')])
worldswin_teams_data = (new_team_data[(new_team_data['WSWin'] == 'Y')]) 

##grouping the teams that won league and that won world series.
league_grouped_teams_data = leaguewin_teams_data.groupby('teamID', as_index = False)[['yearID','LgWin']]
world_grouped_teams_data = worldswin_teams_data.groupby('teamID', as_index = False)[['yearID','WSWin']]

##combining the data sets
combined_world_grouped_teams_data = world_grouped_teams_data.count()
combined_league_grouped_teams_data = league_grouped_teams_data.count()
total_winning_stat = combined_world_grouped_teams_data.merge(combined_league_grouped_teams_data, on='teamID', how ='inner')

##Sorting the data and finding top ten teams of the last 2 decades.
sorted_data = total_winning_stat[['teamID','LgWin','WSWin']].sort_values(['WSWin','LgWin'], ascending=False)
top_ten = sorted_data.iloc[0:10]
top_ten = top_ten.reset_index(drop=True)

print("Top 10 Teams of last 2 decades")
top_ten
```

#### FInding out the top 100 batters and pitchers who played in the top 10 teams 

#### 1. Here also i am considering batsmen and pitchers from the year 1990 till current date. First i am extracting the fields required for finding various batter and pitcher stat required for deciding their performance.

#### 2. Then we merge the new data with the top ten teams data and keep only those players which are present in both tables.By doing this we are assured to get the good players.

#### 3. Next I am taking out the variables we need for our analysis like Batting Average and WHIP. I am applying conditions or range of the variables so that we can get the best players for eg. number of Atbats should be greater than 180 since there are many players who have played 1 or 2 match and their average crosses .400 which should not be the case. After this we group the data by playerid and find the mean figure to get individual performance stat.

#### 4. Then sort the data in descending order to get top 100 players.

```python
##Filtering the data for taking batter stats from the last 2 decades
new_batter_data = batter_data[(batter_data['yearID'] >= 1990)]

##Extracting fields required for taking out batter stats. 
new_batter_data = new_batter_data[['playerID','yearID','teamID','G','AB','R','H','HR','RBI','SO']]


##Merging top 10 teams data with player data so that we get the players linked to those teams
merged_batter_data = new_batter_data.merge(top_ten, on='teamID', how = 'inner')

#Adding a new field Batting average for each batter and dropping NaN values if found
merged_batter_data['Batting Average'] = merged_batter_data['H']/merged_batter_data['AB'] *1.0
merged_batter_data = merged_batter_data.dropna()


##Chosing the minimum number  of AtBats to be 180 since there are cases where hits and Atbats are vey less and their batting 
##average crosses .400 which should not happen since their hits are less than 100.
merged_batter_data = merged_batter_data[(merged_batter_data['AB']>=180) & (merged_batter_data['Batting Average'] >= 0.300)]
merged_batter_data = merged_batter_data.groupby('playerID' , as_index=False).mean()[['playerID','Batting Average','HR','RBI','R']]
batter_sorted_data = merged_batter_data[['playerID','Batting Average','HR','RBI','R']].sort_values(['RBI','Batting Average','HR','R'], ascending=False)

#rounding off data to take out exact figures for each field
batter_sorted_data = batter_sorted_data.round(3)
batter_sorted_data[['HR','R']] = np.round(batter_sorted_data[['HR','R']],0)

#taking out top 100 batters of the last 2 decades
batter_top100_data = batter_sorted_data.iloc[0:100]
batter_top100_data = batter_top100_data.reset_index(drop=True)

print("Data showing top 100 batter with their IDs and stats")
batter_top100_data
```

```python
##Filtering the data for taking pitcher stats from the last 2 decades
new_pitcher_data = pitching_data[(pitching_data['yearID'] >= 1990)]

##Extracting fields required for taking out pitcher stats. 
new_pitcher_data = new_pitcher_data[['playerID','yearID','teamID','W','L','IPouts','H','BB','SO','ERA']]


##Merging top 10 teams data with player data so that we get the players linked to those teams
merged_pitcher_data = new_pitcher_data.merge(top_ten, on='teamID', how = 'inner')

#Adding a new field WHIP for each pitcher and dropping NaN values if found
merged_pitcher_data['WHIP'] = (merged_pitcher_data['BB'] + merged_pitcher_data['H'])*3/merged_pitcher_data['IPouts']
merged_pitcher_data = merged_pitcher_data.dropna()

##Chosing the range of ERA to be [3.00,4.00] for considering the best pitchers.And to check if the number of wins 
##are more than number of losses for the pitcher
##Also checking if WHIP is closer to one so i have taken a range of [0.5,1.5]
merged_pitcher_data = merged_pitcher_data[(merged_pitcher_data['ERA'] >=3.00) & (merged_pitcher_data['ERA'] <= 4.00) 
                                         & (merged_pitcher_data['W'] > merged_pitcher_data['L'] )
                                         & (merged_pitcher_data['WHIP'] >=0.500) & (merged_pitcher_data['WHIP'] <= 1.500)]

merged_pitcher_data = merged_pitcher_data.groupby('playerID',as_index=False).mean()[['playerID','WHIP','SO','ERA','W','L']]

pitcher_sorted_data = merged_pitcher_data[['playerID','WHIP','SO','ERA','W','L']].sort_values(['SO','WHIP','ERA'],ascending=False)

#rounding off data to take out exact figures for each field
pitcher_sorted_data[['W','L','SO']]  = pitcher_sorted_data[['W','L','SO']] .round()
pitcher_sorted_data[['WHIP','ERA']] = pitcher_sorted_data[['WHIP','ERA']].round(3)

#taking out top 100 pitchers of the last 2 decades
pitcher_top100_data = pitcher_sorted_data.iloc[0:100]
pitcher_top100_data = pitcher_top100_data.reset_index(drop=True)

print("Data showing top 100 pitcher with their IDs and stats")
pitcher_top100_data
```

### b. What type of players get higher salary than others - a Correlation between salary and performance of players.

For this question we are not considering the top 100 players but all the players playing or retired since 1990. Here we merge the player data with salary data and find out pearson's r to help find the correlation between performance and salary of players. For batters i have considered Runs Batted In(RBI) as the parameter to consider since this gives more accurate stats as compared to batting average. For pitchers i have chosen  Walks plus Hits per inning pitched(WHIP) as the parameter to consider for performance in matches.Below i have done the operations stated and lastly i am plotting the correlation plot to see the trend for both pitchers and batters and if it matches with what is written in wikipedia .

```python
##Defining a procedure to calculate correlation of 2 variables
def correlation(x, y):
    
    std_x = ((x - x.mean())/x.std(ddof=0))
    std_y = ((y - y.mean())/y.std(ddof=0))
    
    return (std_x * std_y).mean()
```

```python
##Merging Batsmen Data from top 10 teams with salary data and finding out the coorelation between salary and Runs batted in 
##by each batsman

batter_salary_data = batter_sorted_data.merge(salary_data, on = 'playerID', how = 'left')
batter_salary_data = batter_salary_data.groupby('playerID',as_index=False).mean()
batter_salary_data['Salary'] =  batter_salary_data['salary'].round()
del batter_salary_data['salary']
del batter_salary_data['yearID']

batting_salary_corr = correlation(batter_salary_data['RBI'], batter_salary_data['Salary'])

##Finding the player with maximum salary
highest_paid_batsman = batter_salary_data.max()
highest_paid_batsman = master_data[master_data['playerID'] == highest_paid_batsman['playerID']]
highest_paid_batsman['Player_Name'] =  highest_paid_batsman['nameFirst'] + ' ' + highest_paid_batsman['nameLast']


print( '{} = {}'.format('correlation between salary and performance of Batsman',batting_salary_corr) )
print('{} = {}'.format('Highest paid Batsman of last 2 decades from the top ten teams' ,highest_paid_batsman.Player_Name.values[0]))
```

### c. Find the players which have played for a team for more than 12 years?

### d. Finding those players who were managers for a team and had played for the same team, this does not include player-managers?

For this question we are changing our constraints of the dataset , previously it was that we will be looking at players who have played after 1989 but now the new constraint is that we need managers who were not player-managers. So for this purpose we will be looking at the whole dataset. First we group each player by their teams and count the times they played for their team. Next we consider those managers who were not player-managers and count their stint in a a team. Finally we merge both the player data and manager data to find those players who were also managers.

We plot the teams and their count of players who were also managers to see which team more often appoints it own player as manager for that team.

```python
##For this question we are considering the whole dataset

u_batter_data = batter_data[['playerID','stint','teamID']]

##Counting the stint of a player in a team
u_batter_data = u_batter_data.groupby(['playerID','teamID'],as_index=False).count()

##Considering those managers who were not player manager
new_manager_data = manager_data[manager_data['plyrMgr'] == 'N']

##Again counting how many times a manager managed a team
new_manager_data= new_manager_data.groupby(['playerID','teamID'],as_index=False).count()[['playerID','teamID','plyrMgr']]

##Merging data with batter data to find all those batsmen who were also managers
batter_manager = pd.merge(new_manager_data, u_batter_data, on=['playerID', 'teamID'], how ='inner')


batter_manager = batter_manager.merge(master_data , on = 'playerID',how ='left')


#Filtering Player Names field 
batter_manager['Player_Name'] =  batter_manager['nameFirst'] + ' ' + batter_manager['nameLast']


batter_manager = batter_manager[['playerID','Player_Name','teamID','plyrMgr']]

batter_manager.sort_values('plyrMgr',ascending=False)

batter_manager['Stint'] = batter_manager['plyrMgr']


batter_manager = batter_manager.sort_values('plyrMgr',ascending=False).reset_index(drop=True)[['playerID','Player_Name','teamID','Stint']]

batter_manager
```

```python
u_pitcher_data = pitching_data[['playerID','stint','teamID']]

##Counting the stint of a player in a team
u_pitcher_data = u_pitcher_data.groupby(['playerID','teamID'],as_index=False).count()

##Considering those managers who were not player manager
new_manager_data = manager_data[manager_data['plyrMgr'] == 'N']

##Againg counting how many times a manager managed a team
new_manager_data= new_manager_data.groupby(['playerID','teamID'],as_index=False).count()[['playerID','teamID','plyrMgr']]

##Merging data with batter data to find all those pitcher who were also managers
pitcher_manager = pd.merge(new_manager_data, u_pitcher_data, on=['playerID', 'teamID'], how ='inner')


pitcher_manager = pitcher_manager.merge(master_data , on = 'playerID',how ='left')


#Filtering Player Names and plyrMgr field 
pitcher_manager['Player_Name'] =  pitcher_manager['nameFirst'] + ' ' + pitcher_manager['nameLast']
pitcher_manager = pitcher_manager[['playerID','Player_Name','teamID','plyrMgr']]
pitcher_manager.sort_values('plyrMgr',ascending=False)
pitcher_manager['Stint'] = pitcher_manager['plyrMgr']


pitcher_manager = pitcher_manager.sort_values('plyrMgr',ascending=False).reset_index(drop=True)[['playerID','Player_Name','teamID','Stint']]

pitcher_manager
```
